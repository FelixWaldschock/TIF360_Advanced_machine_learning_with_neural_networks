```python
def numberOfZerosInQ(table):
    count = 0
    for i in range(len(table)):
        for j in range(len(table[i])):
            if table[i][j] == 0:
                count += 1
    return count

def numberOfVisitedStates(table):
    count = 0
    tmp = np.sum(table, axis=1)
    count = np.nonzero(tmp)[0].shape[0]
    return count
```

Exercise a:


```python
label = "4_1000"
label = str(label)


# load the rewards from the /logfiles
folder = "logfiles/"
rewards = np.load(folder + "rewards_" + str(label) + ".npy")
gameLength = np.load(folder + "gameLengthTracker_" + str(label) + ".npy")
actions = np.load(folder + "actionTracker_" + str(label) + ".npy")
Q = np.load(folder + "Q-table_" + str(label) + ".npy")
```

```python
def computeMovingAverage(rewards):
    rewardsAverage = []
    averageLength = 100
    for i in range(len(rewards)):
        if i < averageLength:
            rewardsAverage.append(np.mean(rewards[:i]))
        else:
            rewardsAverage.append(np.mean(rewards[i-averageLength:i]))
    return rewardsAverage
```

```python
# get rid of the last 10 episodes
rewards = rewards[:-1]
plt.figure(figsize=(6, 4))
plt.plot(rewards, label='rewards')
plt.plot(computeMovingAverage(rewards=rewards), label='rewards average')
plt.plot(gameLength, label='game length')
plt.title('Rewards over time - Simulation: ' + label + " Parameters a)")
plt.xlabel('Episodes')
plt.ylabel('Rewards')
plt.grid('minor')
plt.legend()

print("Number of zeros in Q: ", numberOfZerosInQ(Q))
print("Number of visited states: ", numberOfVisitedStates(Q))
```
