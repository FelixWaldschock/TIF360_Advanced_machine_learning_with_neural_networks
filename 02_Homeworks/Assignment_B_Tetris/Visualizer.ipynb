```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
import json
# in the "/logfiles" directory get the last timestamp


plt.style.use('science')


def readBestRewardValue():
    with open("logfiles/conf/log.json", "r") as file:
        data = json.load(file)
        return str(data["index"])

label = readBestRewardValue()
```

```python
# load the rewards from the /logfiles
folder = "logfiles/"
rewards = np.load(folder + "rewards_" + str(label) + ".npy")
gameLength = np.load(folder + "gameLengthTracker_" + str(label) + ".npy")
actions = np.load(folder + "actionTracker_" + str(label) + ".npy")
```

```python
# get rid of the last 10 episodes
rewards = rewards[:-1]
```

```python
plt.figure(figsize=(6, 4))
plt.plot(rewards, label='rewards')
plt.plot(gameLength, label='game length')
plt.title('Rewards over time - Simulation: ' + label)
plt.xlabel('Episodes')
plt.ylabel('Rewards')
plt.grid('minor')
plt.legend()
```
